```python
import pandas as pd

import plotly.express as px

import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv("Sample - Superstore.csv", encoding = 'latin-1')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

#Converting date column

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of week'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

#Monthly Sales Analysis

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()

```

```python
sales_by_month
```

```python
fig = px.line(sales_by_month,
              x= 'Order Month',
              y= 'Sales',
              title= 'Monthly sales analysis')
fig.show()
            
```

#SALES BY CATEGORY

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
sales_by_category
```

```python
fig = px.pie(
    sales_by_category,  # DataFrame, not a string
    values='Sales',
    names='Category',
    hole=0.1,
    color_discrete_sequence=px.colors.qualitative.Pastel
)

fig.update_traces(
    textposition='inside',
    textinfo='percent+label'
)

fig.update_layout(
    title_text='Sales Analysis by Category',
    title_font=dict(size=24)
)

fig.show()
    
```

#SALES ANALYSIS BY SUB CATEGORY

```python
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
sales_by_subcategory
```

```python
fig = px.bar(sales_by_subcategory,
             x= 'Sub-Category',
             y= 'Sales',
             title= "Sales analysis by sub category")

fig.show()
             
```

#Monthly Profit analysis

```python
Profit_by_Month = data.groupby('Order Month')['Profit'].sum().reset_index()
Profit_by_Month
```

```python
fig = px.line(Profit_by_Month,
              x= 'Order Month',
              y= 'Profit',
              title= 'Monthly Profit Analysis')

fig.show()
              
```

#Profit by Category

```python
Profit_by_Category = data.groupby('Category')['Profit'].sum().reset_index()
Profit_by_Category
```

```python
fig = px.pie(
    Profit_by_Category,
    values='Profit',
    names='Category',
    hole=0.5,
    color_discrete_sequence=px.colors.qualitative.Pastel
)

fig.update_traces(
    textposition='inside',
    textinfo='percent+label'
)

fig.update_layout(
    title_text='Profit by Category',
    title_font=dict(size=24)
)

fig.show()
```

#Profit by Sun Category

```python
Profit_by_Sub_Category = data.groupby('Sub-Category')['Profit'].sum().reset_index()

Profit_by_Sub_Category

fig = px.bar(Profit_by_Sub_Category,
             x= 'Sub-Category',
             y= 'Profit',
             title= "Profit_by_Sub_Category")

fig.show()
```

#Sales And Profit By Customer Segment

```python
Sales_Profit_by_Segment = (
    data.groupby('Segment')
    .agg({'Sales': 'sum', 'Profit': 'sum'})
    .reset_index())

color_palette = colors.qualitative.Pastel

fig = go.Figure()

# Sales Bar
fig.add_trace(
    go.Bar(
        x=Sales_Profit_by_Segment['Segment'],
        y=Sales_Profit_by_Segment['Sales'],
        name='Sales',
        marker_color=color_palette[0]
    ))

# Profit Bar
fig.add_trace(
    go.Bar(
        x=Sales_Profit_by_Segment['Segment'],
        y=Sales_Profit_by_Segment['Profit'],
        name='Profit',
        marker_color=color_palette[1]
    ))

fig.update_layout(
    title='Sales and Profit by Customer Segment',
    xaxis_title='Customer Segment',
    yaxis_title='Amount')

fig.show()
```

#Sales to profit ratio

```python
Sales_Profit_by_Segment = (
    data.groupby('Segment')
    .agg({'Sales': 'sum', 'Profit': 'sum'})
    .reset_index())
Sales_Profit_by_Segment['Sales_to Profit_Ration']= Sales_Profit_by_Segment['Sales']/Sales_Profit_by_Segment['Profit']

print(Sales_Profit_by_Segment[['Segment', 'Sales_to Profit_Ration']])
```
